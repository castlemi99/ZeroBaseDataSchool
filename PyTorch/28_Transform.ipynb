## Transform on Tensor

```python
transforms.Resize([512, 512])(image) # resize
```

```python
transforms.RandomCrop(size = (512, 512))(image) # randomcrop
```

```python
transforms.RandomCrop(size = (512, 512))(image) # randomcrop
```

```python
# 밝기
transforms.ColorJitter(brightness = 1)(image)
```

```python
transforms.Grayscale()(image)
```

```python
transforms.Pad(padding = (20 , 10))(image) # 그림 바깥 테두리에 padding
```

```python
transforms.RandomAffine(degrees = 90)(image) # 랜덤하게 회전
```

```python
# 한 번에
transform_list = [
    transforms.RandomAffine(degrees = 90),
    transforms.Pad((20,20))
]

# 확률에 의해 적용
transforms.RandomApply(transform_list, p = 0.5)(image)
```

```python
# transform_list 중 랜덤하게 한 개만 적용
transforms.RandomChoice(transform_list)(image)
```

```python
transforms.RandomHorizontalFlip(p = 1.)(image) # 반전
```
