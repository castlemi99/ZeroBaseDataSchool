# 워드 클라우드

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open('./data/06_alice.txt').read()
print(text)
```

```python
alice_mask = np.array(Image.open('./data/06_alice_mask.png'))
alice_mask # 이미지를 픽셀값으로 변환시켜 numpy array로 만들어줌
```

```python
# STOPWORDS : 큰 의미가 없는 단어 토큰 모음
stopwords = set(STOPWORDS)
stopwords
```

```python
stopwords.add('said') # alice.txt에서 'said' 단어가 많아 이것도 포함시켜줌
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,8))
plt.imshow(alice_mask, cmap = plt.cm.gray)
```

```python
wc = WordCloud(
    background_color='white', # 배경색
    max_words= 2000,  # 최대 단어 수
    mask=alice_mask,  # 바탕 배경 지정
    stopwords=stopwords # 불필요한 단어
)

wc = wc.generate(text)
wc.words_ # 각 단어들의 발생빈도를 반환해줌
```

원하는 그림에 alice.txt의 워드클라우드를 씌워보자.

```python
plt.figure(figsize = (12,12))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
