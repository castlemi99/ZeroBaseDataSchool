## Human Learning
먼저, 데이터를 보고 세 품종을 어떻게 구분하면 좋을지 생각해보자.

```python
# boxplot(sepal length vs species )
plt.figure(figsize=(12,6))
sns.boxplot(x = 'sepal length (cm)', y = 'species', data = iris_pd, orient='h')
```

- 특별한 관계는 보이지 않는다.
- 특히 5.5 ~ 6.0 사이는 모든 품종이 겹치고 있으므로 sepal length 필드 하나만 가지고선 구분이 불가할 것 같다.
- virginica는 대체로 5.6 ~ 8.0 사이의 sepal length 값을 가지고 있는데 4.9 정도의 위치의 이상치도 확인된다.

```python
# boxplot(sepal width vs species )
plt.figure(figsize=(12,6))
sns.boxplot(x = 'sepal width (cm)', y = 'species', data = iris_pd, orient='h')
```

- 이 또한 세 품종을 구분하기에는 뚜렷한 특징이 없어보인다.

```python
# boxplot(petal length vs species )
plt.figure(figsize=(12,6))
sns.boxplot(x = 'petal length (cm)', y = 'species', data = iris_pd, orient='h')
```

- setosa는 petal length 필드로 확실히 구분할 수 있을 것 같다.


```python
# boxplot(petal width vs species )
plt.figure(figsize=(12,6))
sns.boxplot(x = 'petal width (cm)', y = 'species', data = iris_pd, orient='h')
```

- 이 또한, setosa는 확실히 구분할 수 있을 것 같다.

```python
# pairplot

sns.pairplot(iris_pd, hue = 'species')
```

- pairplot을 그려보았을 때도 petal 특성으로 setosa의 구분은 확실히 할 수 있을 것 같다는 결론을 내렸다.
- 또한, 다른 두 품종도 petal 특성으로 어느정도 구분은 가능할 것 같다.

- Setosa : 대략 petal length 가 2.5보다 작다
- Versicolor : petal length > 2.5 & petal width < 1.6
- Virginica : petal length > 2.5 & petal width > 1.6

기준값들을 모두 다 구했다면 이는 Decision Tree로 표현될 수 있다.

하지만 먼저 아래 질문에 대해 더 깊이 고민해보자.

"petal length의 기준은 확실하지만, petal width의 기준값은 무엇이 최선일까?"

- 지니계수

분할기준 값이 최소값인 그 때의 값으로 구분한다.
분할기준은 직접 코딩할 수도 있지만, 지금은 Frame Work를 통해 직접 코딩하지 않아도 구할 수 있다.

### Scikit Laern

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier() # 괄호 안에 들어갈 옵션들과 하이퍼파라미터들이 많지만 지금은 생략
```

```python
iris.data.shape
```

```python
# petal의 특성만 봐도 되니, petal 필드만 추출해서 학습시켜보자.
# .fit(학습시킬 데이터, 정답지)
iris_tree.fit(iris.data[:, 2:], iris.target)
```

```python
# 학습이 잘 되었는지 데이터를 다시 주고 예측시켜보자
# .predict(예측시킬 데이터)

y_pred_tr = iris_tree.predict(iris.data[:,2:])
y_pred_tr
```

```python
# 정답지(실제값)
iris.target
```

```python
# 정답률 추출
# accuracy_score(실제값, 예측값)
from sklearn.metrics import accuracy_score

accuracy_score(iris.target, y_pred_tr)
```

```python
# 학습된 모델을 시각화하여 확인할 수 있음
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8))
plot_tree(iris_tree, filled= True)
```

## 데이터 나누기

과적합을 방지하기 위해 훈련용 데이터 / 검증용 데이터 / 평가용 데이터로 데이터를 분리시킬 필요가 있다.

데이터를 나눈 후 다시 학습을 시켜보자.

```python
iris = load_iris()
```

```python
# 학습에 이용할 데이터 준비

features = iris.data[:, 2:]
labels = iris.target
```

```python
# 데이터 나누기
# train_test_split(데이터, 타켓 데이터, test_size = 테스트용 데이터 비중, random_state = 난수)
# random_state : 호출할 때마다 동일한 학습/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 값
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=13)
```

```python
X_train.shape, X_test.shape
```

```python
# 데이터를 나누었다면 꼭! 나눠진 데이터별 타겟 데이터들의 비중이 고르게 나눠졌는지 확인해야한다
import numpy as np

np.unique(y_test, return_counts=True)
```

좀 더 효과적인 학습과 검증을 위해, 타겟 데이터의 비중을 맞춰주는 것이 좋다.

이때 사용하는 옵션이 'stratify'이다.

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.2, 
                                                    stratify=labels,
                                                    random_state=13
                                                    )
X_train.shape, X_test.shape
```

```python
np.unique(y_test, return_counts=True)
```

이제 Train 데이터를 대상으로 다시 모델을 만들어보자

```python
# Decision Tree 의 과적합을 방지하기 위한 옵션
# max_depth : 가지의 깊이를 결정(숫자가 작을수록 모델이 단순해짐)

from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,4))
plot_tree(iris_tree, filled= True)
```

```python
# 학습된 모델에 전체 데이터를 다시 예측시켜보고 정확도를 구해보자
y_pred_tr = iris_tree.predict(iris.data[:,2:])
accuracy_score(iris.target, y_pred_tr)
```

일반적인 데이터에도 95%의 성능이 나올까?

```python
# 위 모델을 Test 데이터를 넣어 예측시켜보자
y_pred_tr = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_tr)
```

```python
# Train, Test 데이터를 분류해서 표시하고 모델의 결정경계까지 표시

scatter_highlight_kwargs = {'s' : 150, 'label' : 'Test data', 'alpha' : 0.9}
scatter_kwargs = {'s' : 120, 'edgecolor' : None,'alpha' :0.9}

plt.figure(figsize=(12,8))
plot_decision_regions(X= features, y = labels,
                      X_highlight=X_test, clf = iris_tree,legend = 2, 
                      scatter_highlight_kwargs=scatter_highlight_kwargs,
                      scatter_kwargs=scatter_kwargs,
                      contour_kwargs={'alpha' : 0.2})
plt.show()
```

## zip 과 언패킹

```python
# 모델을 결정하는데 필요한 중요 features
iris_tree.feature_importances_
```

```python
iris_clf_model = dict(zip(iris.feature_names, iris_tree.feature_importances_))
iris_clf_model
```

```python
list1 = ['a', 'b', 'c']
list2 = [1,2,3]

pairs = [pair for pair in zip(list1, list2)] # 리스트를 튜플로 엮어줄 수 있다.
pairs
```

```python
dict(pairs) #이를 다시 dict형태로 바꿀 수 있다.
```

```python
# 언패킹
x, y = zip(*pairs)
x,y # 튜플 형태로 나옴
```

```python
list(x), list(y) # 리스트 형태로 변환시킬 수 있다.
```
